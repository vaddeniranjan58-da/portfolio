```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv(r"C:\Users\mahes\Downloads\HR-Employee-Attrition.csv")
pd.set_option('display.max_columns',None)
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
#Binary values
data['Attrition']=data['Attrition'].apply(lambda x: 1 if x=='Yes' else 0)
data['Gender']=data['Gender'].apply(lambda x: 1 if x=='Male' else 0)
data['Over18']=data['Over18'].apply(lambda x: 1 if x=='Y' else 0)
data['OverTime']=data['OverTime'].apply(lambda x: 1 if x=='Yes' else 0)
```

```python
data.head()
```

```python
# One-Hot Encoding

data = data.join(
    pd.get_dummies(data['BusinessTravel'], prefix='BusinessTravel')
).drop('BusinessTravel', axis=1)

data = data.join(
    pd.get_dummies(data['Department'], prefix='Department')
).drop('Department', axis=1)

data = data.join(
    pd.get_dummies(data['EducationField'], prefix='Education')
).drop('EducationField', axis=1)

data = data.join(
    pd.get_dummies(data['JobRole'], prefix='JobRole')
).drop('JobRole', axis=1)

data = data.join(
    pd.get_dummies(data['MaritalStatus'], prefix='MaritalStatus')
).drop('MaritalStatus', axis=1)
```

```python
data
```

```python
data=data.map(lambda x:1 if x is True else 0 if x is False else x)
```

```python
data
```

```python
data=data.drop('EmployeeNumber',axis=1)
```

```python
import matplotlib.pyplot as plt
data.hist(figsize=(20,25))
plt.tight_layout()
plt.show()
```

```python
data=data.drop(['EmployeeCount','Over18','StandardHours'],axis=1)
```

data

### Model Training

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
X,y=data.drop('Attrition',axis=1),data['Attrition']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
model = RandomForestClassifier(n_jobs=-1)
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
sorted_importances=dict(sorted(zip(model.feature_names_in_,model.feature_importances_),key=lambda x:x[1],reverse=True))
```

```python
plt.figure(figsize=(20,8))
plt.bar(sorted_importances.keys(), sorted_importances.values())
plt.xticks(rotation=45,ha='right')
plt.show()
```
